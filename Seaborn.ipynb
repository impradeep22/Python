```python
# Distribution plot -- Plot graph with variables  
tips = sns.load_dataset('tips') # data import
```

```python
tips.head()
```

```python
sns.residplot(x = 'total_bill', y = 'tip', data = tips, lowess=True) # residal plot
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips) # by default distribution type or kind is scattered
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex', color = 'green') # change kind value to hex
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde', color = 'green') # change kind value to kde
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg', color = 'green') # change kind value to reg(regression)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'resid', color = 'green') 
```

```python
#Pair plot
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'smoker')
```

```python
x = sns.pairplot(tips, hue = 'smoker')
x.map_diag(sns.distplot)
x.map_upper(sns.kdeplot)
x.map_upper(sns.rugplot)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Matrix plot
```

```python
tips2 = tips.corr()
sns.heatmap(tips2, annot = True, lw = 2, linecolor = '#fff',cmap = 'cool')
```

```python
flights2 = flights.corr()
sns.heatmap(flights2, annot = True, lw = 4, linecolor = '#fff')
```

```python
flights3 = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
sns.heatmap(flights3, lw = 1, linecolor = '#fff', cmap = 'cool')
```

```python
sns.clustermap(flights3, lw = 1, linecolor = '#fff', cmap = 'cool')
```
