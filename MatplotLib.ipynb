```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x =  np.linspace(0, 10,20)
y = x * x
z = x+y
```

```python
x
```

```python
#Plot multiple graphs using single command
plt.plot(x,y,x,x,y,z,y,x)
```

```python
#Plot graph and lable the graph
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X vs Y')
```

```python
plt.plot(x,y)
plt.plot(y,z)
```

```python
#Subplot graph examples
#plt.subplot(rows,columns,)
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(y,z)
plt.subplot(3,2,5)
plt.plot(1/y,z*z)
plt.subplot(3,2,6)
plt.plot(x*y,1/x*z)
plt.tight_layout() #place the graph properly
```

```python
#Type of plot--commonly used

plt.subplot(3,2,1)
plt.plot(x,y)#--Simple plot
plt.subplot(3,2,2)
plt.scatter(x,y)#--scatter plot
plt.subplot(3,2,3)
plt.bar(x,y)#--bar plot
plt.subplot(3,2,4)
plt.hist(x,y)#--hist plot
plt.subplot(3,2,5)
plt.fill(x,y)#--fill plot
plt.tight_layout() #place the graph properly
```

```python
plt.polar(x,y)
```

```python
plt.plot(y,y,label = 'First-example')
plt.legend() #--automatically pickup place the graph label at empty place
plt.legend(loc = 4) #--You can also defined the location by of legend by loc = 1-10(10 possitions ar possible)
```

#### 

```python
#Object orinted or custom plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,2])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Graph')
axes.plot(x,y,x,z)
```

```python
#Multiple custom plot
fig = plt.figure()
axes = fig.add_axes([0,0,2,2])
axes.plot(x,y)
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Graph')
axes2 = fig.add_axes([0.2,0.9,1,1])
axes2.plot(y,x)
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('Graph2')
axes3 = fig.add_axes([2,2,1,1])
axes3.plot(x,y*y)
axes3.set_xlabel('x-axis')
axes3.set_ylabel('y-axis')
axes3.set_title('Graph3')
```

```python
#Subplots using Object Oriented
fig, axes =  plt.subplots(3,3)
axes[(2,2)].plot(x,y)
axes[(0,0)].plot(x,y)
axes[(1,1)].plot(x,y)
axes[(1,2)].plot(x,y)
axes[(2,2)].set_xlabel('x-axis')
axes[(2,2)].set_ylabel('y-axis')
axes[(2,2)].set_title('Graph')
fig.tight_layout()
```

```python
#Defining size of the figure
fig = plt.figure(figsize = (5,5))
axes= fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
#Defining size of the figure in subplots
fig, axes =  plt.subplots(3,3, figsize = (8,8))
axes[(2,2)].plot(x,y)
axes[(0,0)].plot(x,y,color = 'r',lw = 3, linestyle = '--', alpha = .2)
axes[(2,2)].set_xlabel('x-axis')
axes[(2,2)].set_ylabel('y-axis')
axes[(2,2)].set_title('Graph')
fig.tight_layout()
```

```python
# plot Styling
fig = plt.figure(figsize = (5,5))
axes= fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'r',lw = 3, linestyle = '--', alpha = .2)
```

```python
# Markup
fig = plt.figure(figsize = (5,5))
axes= fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'r',lw = 3, linestyle = '--', alpha = .8, marker = 'o',markersize = 10, markerfacecolor = 'b'
          ,markeredgecolor = 'g', markeredgewidth = 3)
```

```python
# Limits
fig = plt.figure(figsize = (5,5))
axes= fig.add_axes([0,0,1,1])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Graph')
axes.set_xlim([0,6])
axes.set_ylim([0,20])
axes.plot(x,y,color = 'r',lw = 3, linestyle = '--', alpha = .8, marker = 'o',markersize = 10, markerfacecolor = 'b'
          ,markeredgecolor = 'g', markeredgewidth = 3)
```
