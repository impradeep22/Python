```python
import numpy as np
import pandas as pd
%matplotlib inline
import plotly as pl
import cufflinks as cf
import plotly.offline as po
po.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.rand(100,5),columns = ['a','b','c','d','e'])
```

```python
df
```

```python
df1 = pd.DataFrame({'x' : ['a','b','c','d','e'], 'y': [1,2,3,4,5],'z': [6,7,8,4,3] })
```

```python
df1
```

```python
#Point Plot
```

```python
df.iplot(kind = 'scatter',x = 'a',y = 'b', mode = 'markers',size = 7,symbol = 'x', colorscale = 'paired')
```

```python
df.scatter_matrix()
```

```python
#surface plot
df2 = pd.DataFrame({'x' : [5,2,3,9,1], 'y': [1,2,3,4,5],'z': [6,7,8,4,3] })
```

```python
df2.iplot(kind = 'surface')
```

```python
cf.datagen.sinwave(10,.25).iplot(kind = 'surface',theme='solar',colorscale='brbg', margin=(0,0,0,0))
```
